#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x_square= np.square(x)
x_exp= np.exp(x)
fig, ax = plt.subplots(nrows=1, ncols =2, figsize=(15,5))
ax1.plot(x, x_square)
ax1.set_title('y= x**2')
ax1.set_xlabel('x')
ax1.set_ylabel('x**2')
ax2.plot(x,x_exp)
ax2.set_title('y=e^x')
ax2.set_xlabel('x')
ax2.set_ylabel('e^x')
ax2.set_yscale("log")
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x_square= np.square(x)
x_exp= np.exp(x)
fig, ax = plt.subplots(nrows=1, ncols =2, figsize=(15,5))
ax1.plot(x, x_square)
ax1.set_title('y= x**2')
ax1.set_xlabel('x')
ax1.set_ylabel('x**2')
ax1.legend()
ax2.plot(x,x_exp)
ax2.set_title('y=e^x')
ax2.set_xlabel('x')
ax2.set_ylabel('e^x')
ax2.set_yscale("log")
ax2.legend()
```
